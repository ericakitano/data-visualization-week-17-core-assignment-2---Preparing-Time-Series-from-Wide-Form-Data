# Import Library

```python
import pandas as pd
```

# **Part 1**

## Load Data

```python
filename = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vQN3Ksa9szQuO4G1-msXWAp17KtVHQCBnuEieu_auu1wSiBf3-krHusIx5VBMkihxj-KZLBosDIGEyR/pub?output=csv'
df = pd.read_csv(filename)
df.head()
```

```python
df.info()
```

## Filter the 4 largest cities into a new dataframe

```python
#create a copy of the dataframe
df2 = df.copy()
```

```python
#create filter
large_cities_filter = (df2['SizeRank'] >= 0) & (df2['SizeRank'] <=3)
```

```python
#check filter
df2[large_cities_filter]
```

```python
#apply filter
df2 = df2[large_cities_filter]
```

```python
df2.head(10)
```

## Melt the data to long-form and prepare it for time series analysis

```python
df2.columns
```

```python
melted = pd.melt(df2.drop(columns=['RegionID','StateName']),
                 id_vars=['SizeRank', 'RegionName','RegionType','State','Metro','CountyName'],
                value_name='Home Value',var_name="Date")
melted
```

### Convert the melted dates into datetime datatype.

```python
melted['Date'] = pd.to_datetime(melted['Date'])
melted.info()
```

### Make the datetime column the index.

```python
melted = melted.set_index("Date")
melted
```

## Resample the dataframe as monthly frequency, grouped by City.

```python
df2_ts = melted.groupby('RegionName').resample("M").sum()
df2_ts
```

# **Part 2**

```python
cities = ['New York','Los Angeles','Houston', 'Chicago']
plot_df = df2_ts.loc[cities].copy()
plot_df
```

```python
plot_df = plot_df.drop(columns='SizeRank')
plot_df
```

```python
unstacked = plot_df.unstack(level=0)
unstacked
```

```python
ax = unstacked['Home Value'].plot()
ax.set_title('Home Value for 4 largest cities between 2000 - 2022')
```
